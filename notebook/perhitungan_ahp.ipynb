## Take the first element if a column has multiple value (list)

```python
def hapus_karakter(x):
    if isinstance(x, str):
        return x.replace("[", "").replace("]", "").replace("'","")
    else:
        return None

def ambil_depan(x):
    x = x.split(',')[0].strip()
    return x

df['jenis tanah'] = df['jenis tanah'].apply(hapus_karakter)
df['jenis tanah'] = df['jenis tanah'].apply(ambil_depan)

df['tekstur tanah'] = df['tekstur tanah'].apply(hapus_karakter)
df['tekstur tanah'] = df['tekstur tanah'].apply(ambil_depan)
```

## Compute GAP/Bobot Jenis Tanah

```python
data_test = df[:5]
data_test
```

```python
df_profil_ideal_tanaman
```

```python
list_gap_jt = []

for i in range(len(data_test)):
    temp = df_profil_ideal_tanaman[['jenis tanaman','jenis tanah', 'jenis tanah ev']]

    # ambil jenis tanah & jenis tanah ev 
    temp2 = data_test.iloc[i, [5, 10]].to_frame().T
    temp2 = temp2.rename(columns={'jenis tanah': 'data lahan', 'jenis tanah ev':'jenis tanah ev test'})  
    
    gap_jt = pd.concat([temp, temp2], axis=1)
   
    gap_jt['data lahan'] = gap_jt['data lahan'].fillna(gap_jt['data lahan'][i])    
    gap_jt['jenis tanah ev test'] = gap_jt['jenis tanah ev test'].fillna(gap_jt['jenis tanah ev test'][i])

    gap_jt['gap_jt'] = gap_jt['jenis tanah ev test'] - gap_jt['jenis tanah ev']
    
    list_gap_jt.append(gap_jt)
```

```python
list_gap_jt[4]
```

```python
def interpolasi_gap(x):
    if x == 0:
        x = 11
    elif x == 10 or x == -10:
        x = 10.5 if x == 10 else 10
    elif x == 20 or x == -20:
        x = 9.5 if x == 20 else 9
    elif x == 30 or x == -30:
        x = 8.5 if x == 30 else 8
    elif x == 40 or x == -40:
        x = 7.5 if x == 40 else 7
    elif x == 50 or x == -50:
        x = 6.5 if x == 50 else 6
    elif x == 60 or x == -60:
        x = 5.5 if x == 60 else 5
    elif x == 70 or x == -70:
        x = 4.5 if x == 70 else 4


    elif x < 0 and x >= -10:
        x = ((11 - 10) * (x - (-10)) / (0-(-10))) + 10
    elif x > 0 and x <= 10:
        x = ((11 - 10.5) * (x - (0)) / (0-(-10))) + 10.5
        
    return x
```

```python
list_bobot_jt = []
for i in range(len(data_test)):
    bobot_jt = list_gap_jt[i]    
    bobot_jt['bobot_jt'] = bobot_jt['gap_jt'].apply(interpolasi_gap)
    list_bobot_jt.append(bobot_jt)
list_bobot_jt[4]
```

```python
list_gap_tt = []

for i in range(len(data_test)):
    temp = df_profil_ideal_tanaman[['jenis tanaman','tekstur tanah', 'tekstur tanah ev']]
    
    temp2 = data_test.iloc[i, [6, 11]].to_frame().T
    temp2 = temp2.rename(columns={'tekstur tanah': 'data lahan', 'tekstur tanah ev':'tekstur tanah ev test'})

    gap_tt = pd.concat([temp, temp2], axis=1)

    gap_tt['data lahan'] = gap_tt['data lahan'].fillna(gap_tt['data lahan'][i])
    gap_tt['tekstur tanah ev test'] = gap_tt['tekstur tanah ev test'].fillna(gap_tt['tekstur tanah ev test'][i])

    gap_tt['gap_tt'] = gap_tt['tekstur tanah ev test'] - gap_tt['tekstur tanah ev']
    list_gap_tt.append(gap_tt)
    
list_gap_tt[4]
```

```python
list_bobot_tt = []

for i in range(len(data_test)):
    bobot_tt = list_gap_tt[i]    
    bobot_tt['bobot_tt'] = bobot_tt['gap_tt'].apply(interpolasi_gap)
    list_bobot_tt.append(bobot_tt)
list_bobot_tt[4]
```
